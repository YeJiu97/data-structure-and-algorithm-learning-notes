## 什么是 queue

queue，中文名队列，是一种特殊的线性表，遵循先进先出的原则。



删除队首的结点复杂度是O(n)，添加回来的复杂度是O(1)：



删除队尾的复杂度是O(1)，添加回来的复杂度也是O(1)：



## 代码实现

**Node Class**

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None
```

**构造函数**

```python
class Queue:
    def __init__(self, value):
        new_node = Node(value)
        self.first = new_node
        self.last = new_node
        self.length = 1
    
    def print_queue(self):
        temp = self.first
        while temp is not None:
            print(temp.value)
            temp = temp.next
```

我们来测试一下：

```python
my_queue = Queue(4)
```

```python
my_queue.print_queue()
```

**enqueue**

enqueue实现的功能如下所示：





```python
def enqueue(self):
    # 调用Node类来生成一个新的结点
    new_node = Node(value)
    # 考虑队列为空的情况
    if self.first is None:
        self.first = new_node
        self.lasst = new_node
    else:
        self.last.next = new_node  # 队列的last结点指向新结点
        self.last = new_node  # 让新结点成为队列的新last结点
    self.length += 1  # 修改队列的大小
```

合并代码：

```python
class Queue:
    def __init__(self, value):
        new_node = Node(value)
        self.first = new_node
        self.last = new_node
        self.length = 1
    
    def print_queue(self):
        temp = self.first
        while temp is not None:
            print(temp.value)
            temp = temp.next
            
    def enqueue(self, value):
        # 调用Node类来生成一个新的结点
        new_node = Node(value)
        # 考虑队列为空的情况
        if self.first is None:
            self.first = new_node
            self.lasst = new_node
        else:
            self.last.next = new_node  # 队列的last结点指向新结点
            self.last = new_node  # 让新结点成为队列的新last结点
        self.length += 1  # 修改队列的大小
```

**dequeue**

dequeue实现的功能如下所示：





代码实现：

```python
def dequeue(self):
    # 考虑到队列的长度为0的情况
    if self.length == 0:
        return None
    # 一个临时变量
    temp = self.first
    # 只有一个结点的队列
    if self.length == 1:
        self.first = None
        self.last = None
    # 队列不止一个结点的情况
    else:
        self.first = self.first.next  # 让队列的first结点变成队列的第二个结点
        temp.next = None  # 然temp指向None
    self.length -= 1  # 修改队列的长度
    return temp  # 返回被弹出的结点
```

合并代码：

```python
class Queue:
    def __init__(self, value):
        new_node = Node(value)
        self.first = new_node
        self.last = new_node
        self.length = 1
    
    def print_queue(self):
        temp = self.first
        while temp is not None:
            print(temp.value)
            temp = temp.next
            
    def enqueue(self, value):
        # 调用Node类来生成一个新的结点
        new_node = Node(value)
        # 考虑队列为空的情况
        if self.first is None:
            self.first = new_node
            self.lasst = new_node
        else:
            self.last.next = new_node  # 队列的last结点指向新结点
            self.last = new_node  # 让新结点成为队列的新last结点
        self.length += 1  # 修改队列的大小
        
    def dequeue(self):
        # 考虑到队列的长度为0的情况
        if self.length == 0:
            return None
        # 一个临时变量
        temp = self.first
        # 只有一个结点的队列
        if self.length == 1:
            self.first = None
            self.last = None
        # 队列不止一个结点的情况
        else:
            self.first = self.first.next  # 让队列的first结点变成队列的第二个结点
            temp.next = None  # 然temp指向None
        self.length -= 1  # 修改队列的长度
        return temp  # 返回被弹出的结点
```

测试代码：

```python
my_queue = Queue(1)
my_queue.enqueue(2)
```
