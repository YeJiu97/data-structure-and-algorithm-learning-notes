## 实现双向链表

**Node Class**

首先，我们需要一个创建Node的class：

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None
        self.prev = None
```

**双向链表构造函数**

接着是创建一个双链表的构造函数来启动双链表的类：

```python
class DoublyLinkedList:
    
    # 构造函数
    def __init__(self, value):
        new_node = Node(value)
        self.head = new_node
        self.tail = new_node
        self.length = 1
    
    # 打印函数
    def print_list(self):
        temp = self.head
        while temp is not None:
            print(temp.value)
            temp = temp.next
```

**pop**

```python
# pop 函数
def pop(self):
    # 如果双向链表为空
    if self.length == 0:
        return None  # 直接返回None，因为没有什么可以弹出的
    
    temp = self.tail  # 创建一个临时变量，让这个变量等于双向链表的最后一个结点
    
    # 如果双向链表中只有一个结点
    if self.length == 1:
        self.head = None
        self.tail = None
    else:
        self.tail = self.tail.prev  # 让最后一个结点等于最后第二个结点
        self.tail.next = None  # 最后第二个结点指向None，如此，最后一个结点被链表弹出来（1/2）
        temp.prev = None  # 最后一个结点指向None而不是最后第二个结点，最后一个结点被完全弹出来（2/2）
    
    self.length -= 1  # 双向链表长度减1
    
    # 将被弹出的最后一个结点返回
    return temp
```

**prepend**

```python
def prepend(self, value):
    
    new_node = Node(value)  # 首先调用Node class来创建一个新结点
    
    if self.length == 0:  # 考虑双向链表为空的情况
        self.head = new_node  # 让head指向这个新结点
        self.tail = new_node  # 让tail指向这个新结点
    else:
        new_node.next = self.head  # 让新结点的next指向链表的第一个结点
        self.head.prev = new_node  # 让链表的第一个结点指向新结点
        self.head = new_node  # 让新结点成为链表的第一个结点
    
    self.length += 1
    return True
```

```python
def pop_first(self):
    
    # 空链表的情况
    if self.length == 0:
        return None
    
    # 创建一个临时变量，用来存储链表的第一个结点
    temp = self.head
    
    # 如果链表只有一个结点
    if self.length == 1:
        self.head = None
        self.tail = None
    else:  # 如果有多个结点
        self.head = self.head.next  # 让head指向第一个结点的下一个结点，也就是第二个结点
        self.head.prev = None  # 让第二个结点指向None
        temp.next = None  # temp指向None
    self.length -= 1  # 修改链表的长度
    return temp  # 返回被弹出的第一个结点
```

**get**

```python
def get(self, index):
    if index < 0 or index >= self.length:
        return None
    temp = self.head
    if index < self.length/2:  # 如果在前半部分
        for _ in range(index):
            temp = temp.next
    else:  # 如果在后半部分
        temp = self.tail  # 从最后一个结点开始
        for _ in range(self.length - 1, index, -1):  # 往前遍历，每次步长为-1
            temp = temp.prev
            
    return temp  # 返回被索引的值
```

**insert**

```python
def insert(self, index, value):
    
    if index < 0 or index > self.length:
        return False
    
    if index == 0:
        return self.prepend(value)
    
    if index == self.length:
        return self.append(value)
    
    new_node = Node(value)  # 创建一个新结点
    before = self.get(index - 1)  # before变量，before指向插入位置的前一个结点
    after = before.next  # after变量，after指向插入位置
    
    new_node.prev = before
    new_node.next = after
    before.next = new_node
    after.prev = new_node
    
    self.length += 1
    return True
```

```python
class DoublyLinkedList:
    
    # 构造函数
    def __init__(self, value):
        new_node = Node(value)
        self.head = new_node
        self.tail = new_node
        self.length = 1
    
    # 打印函数
    def print_list(self):
        temp = self.head
        while temp is not None:
            print(temp.value)
            temp = temp.next
    
    # append函数
    def append(self, value):
        new_node = Node(value)
        if self.head is None:
            self.head = new_node
            self.tail = new_node
        else:
            self.tail.next = new_node
            new_node.prev = self.tail
            self.tail = new_node
        self.length += 1
        return True
    
    # pop 函数
    def pop(self):
        # 如果双向链表为空
        if self.length == 0:
            return None  # 直接返回None，因为没有什么可以弹出的

        temp = self.tail  # 创建一个临时变量，让这个变量等于双向链表的最后一个结点

        # 如果双向链表中只有一个结点
        if self.length == 1:
            self.head = None
            self.tail = None
        else:
            self.tail = self.tail.prev  # 让最后一个结点等于最后第二个结点
            self.tail.next = None  # 最后第二个结点指向None，如此，最后一个结点被链表弹出来（1/2）
            temp.prev = None  # 最后一个结点指向None而不是最后第二个结点，最后一个结点被完全弹出来（2/2）

        self.length -= 1  # 双向链表长度减1

        # 将被弹出的最后一个结点返回
        return temp

    def prepend(self, value):

        new_node = Node(value)  # 首先调用Node class来创建一个新结点

        if self.length == 0:  # 考虑双向链表为空的情况
            self.head = new_node  # 让head指向这个新结点
            self.tail = new_node  # 让tail指向这个新结点
        else:
            new_node.next = self.head  # 让新结点的next指向链表的第一个结点
            self.head.prev = new_node  # 让链表的第一个结点指向新结点
            self.head = new_node  # 让新结点成为链表的第一个结点

        self.length += 1
        return True
    
    def pop_first(self):

        # 空链表的情况
        if self.length == 0:
            return None

        # 创建一个临时变量，用来存储链表的第一个结点
        temp = self.head

        # 如果链表只有一个结点
        if self.length == 1:
            self.head = None
            self.tail = None
        else:  # 如果有多个结点
            self.head = self.head.next  # 让head指向第一个结点的下一个结点，也就是第二个结点
            self.head.prev = None  # 让第二个结点指向None
            temp.next = None  # temp指向None
        self.length -= 1  # 修改链表的长度
        return temp  # 返回被弹出的第一个结点
    
    def get(self, index):
        if index < 0 or index >= self.length:
            return None
        temp = self.head
        if index < self.length/2:  # 如果在前半部分
            for _ in range(index):
                temp = temp.next
        else:  # 如果在后半部分
            temp = self.tail  # 从最后一个结点开始
            for _ in range(self.length - 1, index, -1):  # 往前遍历，每次步长为-1
                temp = temp.prev

        return temp  # 返回被索引的值
    
    def set_value(self, index, value):
        temp = self.get(index)  # 调用get函数来获取对应的结点
        if temp:
            temp.value = value  # 直接进行设置即可
            return True
        return False
    
    def insert(self, index, value):

        if index < 0 or index > self.length:
            return False

        if index == 0:
            return self.prepend(value)

        if index == self.length -1:
            return self.append(value)

        new_node = Node(value)  # 创建一个新结点
        before = self.get(index - 1)  # before变量，before指向插入位置的前一个结点
        after = before.next  # after变量，after指向插入位置

        new_node.prev = before
        new_node.next = after
        before.next = new_node
        after.prev = new_node

        self.length += 1
        return True
```

```python
def remove(self, index):
    
    if index < 0 or index >= self.length:
        return None
    
    if index == 0:
        return self.pop_first()
    
    if index == self.length - 1:
        return self.pop()
    
    temp = self.get(index)
    
    temp.next.prev = temp.prev  # 该结点的下一个结点的上一个结点被设置为该结点的上一个结点
    temp.prev.next = temp.next  # 该结点的上一个结点的下一个结点被设置为该结点的下一个结点
    temp.next = None  # 该节点的下一个结点被设置为None
    temp.prev = None  # 该结点的上一个结点被设置为None
    
    self.length -= 1  # 修改长度
    return temp  # 返回
```

```python
class DoublyLinkedList:
    
    # 构造函数
    def __init__(self, value):
        new_node = Node(value)
        self.head = new_node
        self.tail = new_node
        self.length = 1
    
    # 打印函数
    def print_list(self):
        temp = self.head
        while temp is not None:
            print(temp.value)
            temp = temp.next
    
    # append函数
    def append(self, value):
        new_node = Node(value)
        if self.head is None:
            self.head = new_node
            self.tail = new_node
        else:
            self.tail.next = new_node
            new_node.prev = self.tail
            self.tail = new_node
        self.length += 1
        return True
    
    # pop 函数
    def pop(self):
        # 如果双向链表为空
        if self.length == 0:
            return None  # 直接返回None，因为没有什么可以弹出的

        temp = self.tail  # 创建一个临时变量，让这个变量等于双向链表的最后一个结点

        # 如果双向链表中只有一个结点
        if self.length == 1:
            self.head = None
            self.tail = None
        else:
            self.tail = self.tail.prev  # 让最后一个结点等于最后第二个结点
            self.tail.next = None  # 最后第二个结点指向None，如此，最后一个结点被链表弹出来（1/2）
            temp.prev = None  # 最后一个结点指向None而不是最后第二个结点，最后一个结点被完全弹出来（2/2）

        self.length -= 1  # 双向链表长度减1

        # 将被弹出的最后一个结点返回
        return temp

    def prepend(self, value):

        new_node = Node(value)  # 首先调用Node class来创建一个新结点

        if self.length == 0:  # 考虑双向链表为空的情况
            self.head = new_node  # 让head指向这个新结点
            self.tail = new_node  # 让tail指向这个新结点
        else:
            new_node.next = self.head  # 让新结点的next指向链表的第一个结点
            self.head.prev = new_node  # 让链表的第一个结点指向新结点
            self.head = new_node  # 让新结点成为链表的第一个结点

        self.length += 1
        return True
    
    def pop_first(self):

        # 空链表的情况
        if self.length == 0:
            return None

        # 创建一个临时变量，用来存储链表的第一个结点
        temp = self.head

        # 如果链表只有一个结点
        if self.length == 1:
            self.head = None
            self.tail = None
        else:  # 如果有多个结点
            self.head = self.head.next  # 让head指向第一个结点的下一个结点，也就是第二个结点
            self.head.prev = None  # 让第二个结点指向None
            temp.next = None  # temp指向None
        self.length -= 1  # 修改链表的长度
        return temp  # 返回被弹出的第一个结点
    
    def get(self, index):
        if index < 0 or index >= self.length:
            return None
        temp = self.head
        if index < self.length/2:  # 如果在前半部分
            for _ in range(index):
                temp = temp.next
        else:  # 如果在后半部分
            temp = self.tail  # 从最后一个结点开始
            for _ in range(self.length - 1, index, -1):  # 往前遍历，每次步长为-1
                temp = temp.prev

        return temp  # 返回被索引的值
    
    def set_value(self, index, value):
        temp = self.get(index)  # 调用get函数来获取对应的结点
        if temp:
            temp.value = value  # 直接进行设置即可
            return True
        return False
    
    def insert(self, index, value):

        if index < 0 or index > self.length:
            return False

        if index == 0:
            return self.prepend(value)

        if index == self.length -1:
            return self.append(value)

        new_node = Node(value)  # 创建一个新结点
        before = self.get(index - 1)  # before变量，before指向插入位置的前一个结点
        after = before.next  # after变量，after指向插入位置

        new_node.prev = before
        new_node.next = after
        before.next = new_node
        after.prev = new_node

        self.length += 1
        return True
    
    def remove(self, index):

        if index < 0 or index >= self.length:
            return None

        if index == 0:
            return self.pop_first()

        if index == self.length - 1:
            return self.pop()

        temp = self.get(index)

        temp.next.prev = temp.prev  # 该结点的下一个结点的上一个结点被设置为该结点的上一个结点
        temp.prev.next = temp.next  # 该结点的上一个结点的下一个结点被设置为该结点的下一个结点
        temp.next = None  # 该节点的下一个结点被设置为None
        temp.prev = None  # 该结点的上一个结点被设置为None

        self.length -= 1  # 修改长度
        return temp  # 返回
```
