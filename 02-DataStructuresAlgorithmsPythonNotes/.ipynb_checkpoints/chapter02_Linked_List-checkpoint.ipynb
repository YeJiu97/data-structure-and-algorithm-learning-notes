**pop 弹出**

弹出链表的最后一个项目，并且将其返回：



我们需要考虑空链表的情况：



原先的代码可以参考下图：



将最后第二个结点的next指向修改成为None。

很遗憾的是，我们并不能够像数组一样直接弹出最后一个元素，而是需要通过head来不断的追溯，直到最后一个结点。

这里需要引用两个变量：

1. pre
2. temp

来协助进行操作。

逻辑如下所示：

首先将pre和temp指向head：



对于如果temp.next的结果并不是None，那么我们将temp移动到下一个结点：



也就是 temp = temp.next。

然后我们重复操作：





直到：



此时，temp.next的结果为None，就可以在这里停止了。

我们让tail指向pre所在的位置：



而temp现在指向是元向temp的位置：



我们只需要返回temp就好了。

**我们现在可以实现pop的代码了**

```python
def pop(self):
    
    # 创建两个用来保存的临时变量
    temp = self.head
    pre = self.head
    
    if self.length == 0:  # 首先需要考虑长度
        return None  # 如果链表的长度为0,那么直接返回None
    
    while(temp.next):  # 在temp.next不为None的时候进行循环
        pre = temp  # 让 pre 指向 temp现在的结点
        temp = temp.next  # 让temp为temp的下一个结点
    
    # 当上面的循环结束的时候，链表中的所有的结点都被遍历了
    self.tail = pre  # 让链表的tail指向 pre所在的结点
    self.tail.next = None  # 并且让tail指向None
    
    self.length -= 1  # 让链表的长度减1
    
    if self.length == 0:  # 考虑到链表只有一个结点的情况
        self.head = None  # head指向None
        self.tail = None  # tail指向None
    
    return temp  # 我们需要返回被弹出的最后一个结点
```

**prepend**

.prepend() 方法将指定元素插入到匹配元素里面作为它的第一个子元素。



如上图所是，我们有一个新创建的结点，以及一个已经被创建的链表。

我们要做的是让head指向新创建的结点，并且将新创建的结点添加到链表中：



但是我们需要考虑空链表的情况，在这样的情况之下，head和tail都会指向新创建的结点：



```python
def prepend(self, value):
    new_node = Node(value)  # 首先使用Node类来创建一个新的结点
    
    # 接着我们需要考虑链表为空的情况
    if self.length == 0:
        self.head = new_node
        self.tail = new_node
    else:  # 如果不是空链表
        new_node.next = self.head  # 让新结点指向链表的head结点
        self.head = new_node  # 让将链表的head设置为新结点
    
    # 不要忘记对链表的长度增加1
    self.length += 1
    return True
```

**Pop First**

pop()函数会将链表的最后一个结点进行弹出，但是我们需要考虑弹出链表的第一个结点的情况，这个时候我们就需要有一个pop_first()函数来执行这一个功能。

实现的逻辑非常的简单。

首先，创建一个临时变量，使其指向head：



接着让head指向链表的第二个结点：



temp的next需要指向None：



最后不要忘记将链表的长度减去1.

```python
def pop_first(self):
    
    # 如果链表是空链表，直接返回None
    if self.length == 0:
        return None
    
    temp = self.head  # 创建一个临时变量
    self.head = self.head.next  # 让head指向下一个结点
    temp.next = None  # 让被弹出的结点指向None
    self.length -= 1  # 将长度减去1
    
    # 考虑链表只有一个结点的情况
    if self.length == 0:
        self.tail = None
    
    return temp
```

```python
# 我们来对代码进行整合
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None

class LinkedList:
    
    #  初始化函数
    def __init__(self, value):
        new_node = Node(value)  # 调用结点类
        self.head = new_node  # 头 为这个new_node
        self.tail = new_node  # 尾 为这个new_node
        self.length = 1  # 此时的长度为1
    
    # 打印链表中的所有的结点
    def print_list(self):
        temp = self.head  # 创建一个临时变量
    
        while temp is not None:  # 如果不是None
            print(temp.value)  # 打印值
            temp = temp.next  # 让temp指向下一个结点
    
    def append(self, value):
        new_node = Node(value)  # 首先创建一个新的node
        if self.head is None:  # 如果链表中现在没有node
            self.head = new_node
            self.tail = new_node
        else:  # 否则的话
            self.tail.next = new_node  # 让 现在的tail 指向new_node
            self.tail = new_node  # 将 tail 的值修改成 new_node
        self.length += 1  # 对链表的长度进行更新
        return True
    
    def pop(self):

        # 创建两个用来保存的临时变量
        temp = self.head
        pre = self.head

        if self.length == 0:  # 首先需要考虑长度
            return None  # 如果链表的长度为0,那么直接返回None

        while(temp.next):  # 在temp.next不为None的时候进行循环
            pre = temp  # 让 pre 指向 temp现在的结点
            temp = temp.next  # 让temp为temp的下一个结点

        # 当上面的循环结束的时候，链表中的所有的结点都被遍历了
        self.tail = pre  # 让链表的tail指向 pre所在的结点
        self.tail.next = None  # 并且让tail指向None

        self.length -= 1  # 让链表的长度减1

        if self.length == 0:  # 考虑到链表只有一个结点的情况
            self.head = None  # head指向None
            self.tail = None  # tail指向None

        return temp  # 我们需要返回被弹出的最后一个结点


    def prepend(self, value):
        new_node = Node(value)  # 首先使用Node类来创建一个新的结点

        # 接着我们需要考虑链表为空的情况
        if self.length == 0:
            self.head = new_node
            self.tail = new_node
        else:  # 如果不是空链表
            new_node.next = self.head  # 让新结点指向链表的head结点
            self.head = new_node  # 让将链表的head设置为新结点

        # 不要忘记对链表的长度增加1
        self.length += 1
        return True

    def pop_first(self):

        # 如果链表是空链表，直接返回None
        if self.length == 0:
            return None

        temp = self.head  # 创建一个临时变量
        self.head = self.head.next  # 让head指向下一个结点
        temp.next = None  # 让被弹出的结点指向None
        self.length -= 1  # 将长度减去1

        # 考虑链表只有一个结点的情况
        if self.length == 0:
            self.tail = None

        return temp
```

并且set是Python的一个关键字，所以我们使用set_value()。

set的作用在于修改某一个位置的结点的值。

```python
def set_value(self, index, value):
    temp = self.get(index)  # 调用get函数
    if temp:
        temp.value = value
        return True
    return False
```

```python
# 我们来对代码进行整合
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None

class LinkedList:
    
    #  初始化函数
    def __init__(self, value):
        new_node = Node(value)  # 调用结点类
        self.head = new_node  # 头 为这个new_node
        self.tail = new_node  # 尾 为这个new_node
        self.length = 1  # 此时的长度为1
    
    # 打印链表中的所有的结点
    def print_list(self):
        temp = self.head  # 创建一个临时变量
    
        while temp is not None:  # 如果不是None
            print(temp.value)  # 打印值
            temp = temp.next  # 让temp指向下一个结点
    
    def append(self, value):
        new_node = Node(value)  # 首先创建一个新的node
        if self.head is None:  # 如果链表中现在没有node
            self.head = new_node
            self.tail = new_node
        else:  # 否则的话
            self.tail.next = new_node  # 让 现在的tail 指向new_node
            self.tail = new_node  # 将 tail 的值修改成 new_node
        self.length += 1  # 对链表的长度进行更新
        return True
    
    def pop(self):

        # 创建两个用来保存的临时变量
        temp = self.head
        pre = self.head

        if self.length == 0:  # 首先需要考虑长度
            return None  # 如果链表的长度为0,那么直接返回None

        while(temp.next):  # 在temp.next不为None的时候进行循环
            pre = temp  # 让 pre 指向 temp现在的结点
            temp = temp.next  # 让temp为temp的下一个结点

        # 当上面的循环结束的时候，链表中的所有的结点都被遍历了
        self.tail = pre  # 让链表的tail指向 pre所在的结点
        self.tail.next = None  # 并且让tail指向None

        self.length -= 1  # 让链表的长度减1

        if self.length == 0:  # 考虑到链表只有一个结点的情况
            self.head = None  # head指向None
            self.tail = None  # tail指向None

        return temp  # 我们需要返回被弹出的最后一个结点


    def prepend(self, value):
        new_node = Node(value)  # 首先使用Node类来创建一个新的结点

        # 接着我们需要考虑链表为空的情况
        if self.length == 0:
            self.head = new_node
            self.tail = new_node
        else:  # 如果不是空链表
            new_node.next = self.head  # 让新结点指向链表的head结点
            self.head = new_node  # 让将链表的head设置为新结点

        # 不要忘记对链表的长度增加1
        self.length += 1
        return True

    def pop_first(self):

        # 如果链表是空链表，直接返回None
        if self.length == 0:
            return None

        temp = self.head  # 创建一个临时变量
        self.head = self.head.next  # 让head指向下一个结点
        temp.next = None  # 让被弹出的结点指向None
        self.length -= 1  # 将长度减去1

        # 考虑链表只有一个结点的情况
        if self.length == 0:
            self.tail = None

        return temp

    def get(self, index):

        # 需要考虑到这样的两种情况，这两种情况链表并不支持
        if index < 0 or index >= self.length:
            return None

        # 使用一个临时变量
        temp = self.head

        # 接着开始使用for循环来进行遍历
        for _ in range(index):
            temp = temp.next

        # 返回结果
        return temp
    
    def set_value(self, index, value):
        temp = self.get(index)  # 调用get函数
        if temp:
            temp.value = value
            return True
        return False
```

**insert**

insert可以在链表的指定位置添加一个新的结点。

我们不能够使用小于0的索引，或者使用大于链表长度的索引：



如果输入的是0,那么新的结点将会取代原先的head：



由此，我们可以直到，插入的结点会将指定位置的结点和之后的所有的结点往后挪动一个位置。



```python
def insert(self, index, value):
    
    # 首先对索引进行判断
    if index < 0 or index > self.length:
        return False
    # 如果是在最开始，那么等同于在链表的前端进行添加
    if index == 0:
        return self.prepend(value)
    # 如果是在最后，那么就是在链表的最后添加
    if index == self.length:
        return self.append(value)

    # 然后我们需要考虑在链表的中间进行添加的情况
    new_node = Node(value)
    # temp为执行位置的上一个结点
    temp = self.get(index - 1)
    
    # 让新结点指向temp所在的结点的next
    new_node.next = temp.next
    # 让temp的next指向新的结点
    temp.next = new_node
    
    # 修改长度
    self.length += 1
    return True
```

```python
# 我们来对代码进行整合
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None

class LinkedList:
    
    #  初始化函数
    def __init__(self, value):
        new_node = Node(value)  # 调用结点类
        self.head = new_node  # 头 为这个new_node
        self.tail = new_node  # 尾 为这个new_node
        self.length = 1  # 此时的长度为1
    
    # 打印链表中的所有的结点
    def print_list(self):
        temp = self.head  # 创建一个临时变量
    
        while temp is not None:  # 如果不是None
            print(temp.value)  # 打印值
            temp = temp.next  # 让temp指向下一个结点
    
    def append(self, value):
        new_node = Node(value)  # 首先创建一个新的node
        if self.head is None:  # 如果链表中现在没有node
            self.head = new_node
            self.tail = new_node
        else:  # 否则的话
            self.tail.next = new_node  # 让 现在的tail 指向new_node
            self.tail = new_node  # 将 tail 的值修改成 new_node
        self.length += 1  # 对链表的长度进行更新
        return True
    
    def pop(self):

        # 创建两个用来保存的临时变量
        temp = self.head
        pre = self.head

        if self.length == 0:  # 首先需要考虑长度
            return None  # 如果链表的长度为0,那么直接返回None

        while(temp.next):  # 在temp.next不为None的时候进行循环
            pre = temp  # 让 pre 指向 temp现在的结点
            temp = temp.next  # 让temp为temp的下一个结点

        # 当上面的循环结束的时候，链表中的所有的结点都被遍历了
        self.tail = pre  # 让链表的tail指向 pre所在的结点
        self.tail.next = None  # 并且让tail指向None

        self.length -= 1  # 让链表的长度减1

        if self.length == 0:  # 考虑到链表只有一个结点的情况
            self.head = None  # head指向None
            self.tail = None  # tail指向None

        return temp  # 我们需要返回被弹出的最后一个结点


    def prepend(self, value):
        new_node = Node(value)  # 首先使用Node类来创建一个新的结点

        # 接着我们需要考虑链表为空的情况
        if self.length == 0:
            self.head = new_node
            self.tail = new_node
        else:  # 如果不是空链表
            new_node.next = self.head  # 让新结点指向链表的head结点
            self.head = new_node  # 让将链表的head设置为新结点

        # 不要忘记对链表的长度增加1
        self.length += 1
        return True

    def pop_first(self):

        # 如果链表是空链表，直接返回None
        if self.length == 0:
            return None

        temp = self.head  # 创建一个临时变量
        self.head = self.head.next  # 让head指向下一个结点
        temp.next = None  # 让被弹出的结点指向None
        self.length -= 1  # 将长度减去1

        # 考虑链表只有一个结点的情况
        if self.length == 0:
            self.tail = None

        return temp

    def get(self, index):

        # 需要考虑到这样的两种情况，这两种情况链表并不支持
        if index < 0 or index >= self.length:
            return None

        # 使用一个临时变量
        temp = self.head

        # 接着开始使用for循环来进行遍历
        for _ in range(index):
            temp = temp.next

        # 返回结果
        return temp
    
    def set_value(self, index, value):
        temp = self.get(index)  # 调用get函数
        if temp:
            temp.value = value
            return True
        return False

    def insert(self, index, value):

        # 首先对索引进行判断
        if index < 0 or index > self.length:
            return False
        # 如果是在最开始，那么等同于在链表的前端进行添加
        if index == 0:
            return self.prepend(value)
        # 如果是在最后，那么就是在链表的最后添加
        if index == self.length:
            return self.append(value)

        # 然后我们需要考虑在链表的中间进行添加的情况
        new_node = Node(value)
        # temp为执行位置的上一个结点
        temp = self.get(index - 1)

        # 让新结点指向temp所在的结点的next
        new_node.next = temp.next
        # 让temp的next指向新的结点
        temp.next = new_node

        # 修改长度
        self.length += 1
        return True
```

**remove**

remove函数能够移除某一个指定位置的结点。

```python
def remove(self, index):
    
    # 首先我们需要判断index是否是合法的
    if index < 0 or index >= self.length:
        return None
    
    # 如果index是0,那么相当于弹出链表的head
    if index == 0:
        return self.pop_first()
    
    # 同样的逻辑
    if index == self.length - 1:
        return self.pop()
    
    # 创建两个临时变量
    prev = self.get(index - 1)
    temp = prev.next
    
    # 开始进行修改
    prev.next = temp.next
    temp.next = None
    
    self.length -= 1
    return temp.value
```

**reverse**

如何对链表进行翻转，这是一个非常常见的面试问题。

这是最开始的链表：



这是翻转的结果：



我们首先让temp指向head所在的结点，然后切换head和tail：



接着我们需要：



然后进行不断的循环：



```python
def reverse(self):
    
    # 首先将head和tail进行对换
    temp = self.head
    self.head = self.tail
    self.tail = temp
    
    # 创建两个临时变量
    after = temp.next
    before = None
    
    for _ in range(self.length):  # 对每个结点进行循环
        after = temp.next  # after设置为head的下一个结点
        temp.next = before  # 同理
        before = temp
        temp = after
    
```

## 最终代码

```python
# Node 结点类

class Node:
    def __init__(self, value):
        self.value = value  # 将Node的值设置为传入的值
        self.next = None  # 让Node的next指向None
```

```python
# 链表和功能类

class LinkedList:
    
    # 初始化函数
    def __init__(self, value):
        new_node = Node(value)  # 调用Node类创建新的结点
        self.head = new_node  # head指向这个结点
        self.tail = new_node  # tail指向这个结点
        self.length = 1  # 这个时候length的长度只有1
    
    # 打印所有的结点
    def print_list(self):
        temp = self.head  # 创建一个临时变量
        
        while temp is not None:  # 当temp不为None的时候
            print(temp.value)  # 打印temp的值
            temp = temp.next  # 让temp等下一个结点
        
    # 在链表的结尾添加一个Node
    def append(self, value):
        new_node = Node(value)  # 首先需要调用Node class来创建结点
        if self.head is None:  # 考虑到链表现在为空的情况
            self.head = new_node
            self.tail = new_node
        else:  # 否则的话
            self.tail.next = new_node  # 让现在的tail的值指向新结点
            self.tail = new_node  # 让链表的tail设置为新结点
        self.length += 1  # 对链表的长度进行更新
        return True
    
    # 弹出链表的最后一个结点
    def pop(self):
        # 创建两个用来保存的临时变量
        temp = self.head
        pre = self.head
        
        if self.length == 0:  # 如果链表的长度为0
            return None  # 直接返回None
        
        while (temp.next):  # 当temp.next不为None的时候进行循环
            pre = temp  # 让pre指向temp现在的结点
            temp = temp.next  # 让temp指向temp的下一个结点
        
        # 上面循环结束，这个时候就是链表被遍历万了
        self.tail = pre  # 链表的tail指向pre所在的结点，也就是最后第二个结点
        self.tail.next = None  # 最后一个结点被弹出了，所以最后第二个结点指向None
        
        self.length -= 1  # 对链表的长度进行修改
        
        if self.length == 0:  # 考虑到链表为空的情况，也就是原先为1
            self.head = None
            self.tail = None
        
        return temp  # 需要将链表被弹出的最后一个结点返回
    
    # 在链表的开始添加结点
    def prepend(self, value):
        # 调用Node class来创建一个新结点
        new_node = Node(value)
        # 考虑空链表的情况
        if self.length == 0:
            self.head = new_node
            self.tail = new_node
        else:  #如果不是空链表
            new_node.next = self.head  # 让新结点指向原先链表的head
            self.head = new_node  # 将新结点设置为链表的head
        # 对链表的长度进行修改
        self.length += 1
        return True
    
    # 弹出链表的第一个结点
    def pop_first(self):
        # 先考虑空链表的情况
        if self.length == 0:
            return None
        
        temp = self.head  # 创建一个临时变量，使其为链表的第一个结点head
        self.head = self.head.next  # 让链表的head变成第二个结点
        temp.next = None  # 原先的第一个结点指向None
        self.length -= 1  # 让链表的长度减1
        
        # 考虑链表只有一个结点的情况
        if self.length == 0:
            self.tail = None
        
        return temp
    
    # 获得结点
    def get(self, index):
        # 考虑给出来的索引是不合法的情况
        if index < 0 or index >= self.length:
            return None
        
        # 使用一个临时变量
        temp = self.head
        
        # 接着开始使用for来进行循环
        for _ in range(index):
            temp = temp.next
        
        # 返回结果既可
        return temp
    
    # 设置已有结点的值
    def set_value(self, index, value):
        temp = self.get(index)  # 调用get函数
        if temp:  # 如果这个位置存在着结点的话
            temp.value = value  # 将其设置为需要的数值
            return True  # 返回真
        return False  # 如果这个位置不存在着结点，if不会被执行
    
    # 插入一个结点
    def insert(self, index, value):
        # 首先对索引值进行判断
        if index < 0 or index > self.length:
            return False
        # 如果是在链表的最开始插入一个结点
        if index == 0:
            return self.prepend(value)  # 等同于对链表进行结点添加
        # 同理，如果是在链表的最后插入一个结点
        if index == self.length:
            return self.append(value)
        # 接着我们考虑在链表的中间插入结点的情况
        new_node = Node(value)
        # 将temp设置为插入位置的结点的上一个结点
        temp = self.get(index - 1)
        
        # 让新结点指向temp指向原先链表的该位置的结点
        new_node.next = temp.next
        # 让temp指向新结点而不是原先的结点
        temp.next = new_node
        
        # 修改长度
        self.length += 1
        return True
    
    # 移除指定的结点
    def remove(self, index):
        # 首先判断index是否合法
        if index < 0 or index >= self.length:
            return None
        # 如果index是0,那么相当于弹出链表的第一个结点
        if index == 0:
            return self.pop_first()
        # 同理
        if index == self.length - 1:
            return self.pop()
        # 创建两个临时变量
        prev = self.get(index - 1)
        temp = prev.next
        
        # 开始进行修改
        prev.next = temp.next
        temp.next = None
        
        # 修改长度
        self.length -= 1
        
        #返回
        return temp.value
    
    # 开始进行翻转
    def reverse(self):
        # 首先让head和tail进行对换
        temp = self.head
        self.head = self.tail
        self.tail = temp
        # 创建两个临时变量
        after = temp.next
        before = None
        
        for _ in range(self.length):
            after = temp.next
            temp.next = before
            before = temp
            temp = after
```
