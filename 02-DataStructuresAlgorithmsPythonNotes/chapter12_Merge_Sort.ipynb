## Merge Code

```python
def merge(list1, list2):
    
    combined = []
    i = 0
    j = 0
    
    while i < len(list1) and j < len(list2):  # 一旦有一个列表被完全确认，就停止对两个列表的共同确认
        if list1[i] < list2[j]:
            combined.append(list1[i])
            i += 1
        else:
            combined.append(list2[j])
            j += 1
    
    # 对剩下的那个列表中的元素进行添加
    while i < len(list1):
        combined.append(list1[i])
        i += 1
    while j < len(list2):
        combined.append(list2[j])
        j += 1
    
    return combined
```

## Sort Code

```python
def merge_sort(my_list):
    
    # 将传入的列表分成两个子列表
    if len(my_list) == 1:
        return my_list
    mid = int(len(my_list)/2)
    left = my_list[:mid]
    right = my_list[mid:]
    
    # 调用merge函数，并且返回结果
    return merge(merge_sort(left), merge_sort(right))  # 进行递归
```
