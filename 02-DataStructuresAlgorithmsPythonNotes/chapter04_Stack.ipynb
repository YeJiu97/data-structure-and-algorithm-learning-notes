## 什么是 stack

stack是一种数据的存放方式，特点为 Last In First Out（LIFO），如下图所示：





堆栈（stack）的复杂度为O(n)。

## 代码

**constructor**

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None
```

```python
class Stack:
    
    def __init__(self, value):
        new_node = Node(value)
        self.top = new_node
        self.height = 1
    
    def print_stack(self):
        temp = self.top
        while temp is not None:
            print(temp.value)
            temp = temp.next
```

```python
my_stack = Stack(4)

my_stack.print_stack()
```

**push**

```python
def push(self, value):
    new_node = Node(value)  # 首先调用
    if self.height == 0:  # 考虑到堆栈为空的情况
        self.top = new_node  # 让top指向新结点
    else:
        new_node.next = self.top  # 让新结点指向原先的堆栈的top结点
        self.top = new_node  # 让堆栈的top结点为新结点
    self.height += 1  # 修改结点的长度
```

结合在一起：

```python
class Stack:
    
    def __init__(self, value):
        new_node = Node(value)
        self.top = new_node
        self.height = 1
    
    def print_stack(self):
        temp = self.top
        while temp is not None:
            print(temp.value)
            temp = temp.next
    
    def push(self, value):
        new_node = Node(value)  # 首先调用
        if self.height == 0:  # 考虑到堆栈为空的情况
            self.top = new_node  # 让top指向新结点
        else:
            new_node.next = self.top  # 让新结点指向原先的堆栈的top结点
            self.top = new_node  # 让堆栈的top结点为新结点
        self.height += 1  # 修改结点的长度
```

测试一下：

```python
my_stack = Stack(2)
my_stack.push(1)

my_stack.print_stack()
```

**pop**

pop将会弹出和返回堆栈（stack）的顶端（top）结点：



```python
def pop(self):
    
    # 考虑到堆栈为空的情况
    if self.height == 0:
        return None  # 直接返回None
    temp = self.top  # 让临时变量等于堆栈的top结点
    self.top = self.top.next  # 将top结点修改为top结点的下一个结点
    temp.next = None  # 将temp的next修改为None
    
    self.height -= 1  # 修改堆栈的长度
    
    return temp  # 返回temp
```

合并到Stack类中：

```python
class Stack:
    
    def __init__(self, value):
        new_node = Node(value)
        self.top = new_node
        self.height = 1
    
    def print_stack(self):
        temp = self.top
        while temp is not None:
            print(temp.value)
            temp = temp.next
    
    def push(self, value):
        new_node = Node(value)  # 首先调用
        if self.height == 0:  # 考虑到堆栈为空的情况
            self.top = new_node  # 让top指向新结点
        else:
            new_node.next = self.top  # 让新结点指向原先的堆栈的top结点
            self.top = new_node  # 让堆栈的top结点为新结点
        self.height += 1  # 修改结点的长度
        
    def pop(self):

        # 考虑到堆栈为空的情况
        if self.height == 0:
            return None  # 直接返回None
        temp = self.top  # 让临时变量等于堆栈的top结点
        self.top = self.top.next  # 将top结点修改为top结点的下一个结点
        temp.next = None  # 将temp的next修改为None

        self.height -= 1  # 修改堆栈的长度

        return temp  # 返回temp
```

测试一下：

```python
my_stack = Stack(1)

my_stack.push(2)
my_stack.push(3)
my_stack.push(4)
```

```python
my_stack.print_stack()
```
