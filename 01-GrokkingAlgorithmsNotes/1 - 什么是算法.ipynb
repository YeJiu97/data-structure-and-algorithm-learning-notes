```python
def binary_search(list, item):
    low = 0  # Python0
    high = len(list) -1  # len()list0-1
    
    while low <= high:  # 
        
        mid = (low + high)  #  
        
        guess = list[mid]  # 
        if guess == item:  # 
            return mid  # 
        if guess > item:  # 
            high = mid -1  # mid
        else:  # mid
            low = mid + 1
            
    return None
```

```python
# 

testing_list = [6, 7, 8, 9, 10]  # 

print(binary_search(testing_list, 7))  # 0
```

```python
print(binary_search(testing_list, -1))
```
